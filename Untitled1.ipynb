```python
m = 6
for n in range(0, m + 1):
    if n % 2 == 0:
      print(n, "es par y su doble es", n * 2)
    else:
      print(n, "es impar y su triple es", n * 3)
```

```python
def reconocerparimpar(m):
  try:
    m = float(m)
    if m.is_integer() and m >= 0:
      for n in range(0, int(m + 1)):
        if n % 2 == 0:
          print(n, "es par y su doble es", n * 2)
        else:
          print(n, "es impar y su triple es", n * 3)
    elif m.is_integer() and m < 0:
        print("El número es negativo")
    else:
        print("El número no es entero")
  except ValueError:
    print("El valor ingresado no es un número")
```

```python
def contador_vocales1(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  numeros = "0123456789"
  espacio = " "
  contador = 0
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0

  for letra in palabra.lower():
    if letra in vocales:
      contador += 1
    elif letra in consonantes:
      contador1 += 1
    elif letra in numeros:
      contador2 += 1
    elif letra in espacio:
      contador3 += 1
    else:
      contador4 += 1
  return({"Cantidad de vocales:": contador,
          "Cantidad de consonates:": contador1,
          "Cantidad de números:": contador2,
          "Cantidad de espacios:": contador3,
          "Cantidad de otros caracteres:": contador4})
```

```python
def producto_cruz(v1, v2):
 if len(v1) != 3 or len(v2) != 3:
  return "Los vectores deben ser de 3 dimensiones"
 else:
  resultado = [
 v1[1]*v2[2] - v1[2]*v2[1],
 v1[2]*v2[0] - v1[0]*v2[2],
 v1[0]*v2[1] - v1[1]*v2[0]
 ]
 return resultado
```
