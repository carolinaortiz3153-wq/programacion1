```python
def productopunto_vectores(v1, v2):
  return sum(a*b for a, b in zip(v1, v2))

# Programa principal
try:
    v1 = list(map(float, input("Ingrese el primer vector separado por espacios: ").split()))
    v2 = list(map(float, input("Ingrese el segundo vector separado por espacios: ").split()))

    if len(v1) == len(v2):
        print("Suma de vectores:", productopunto_vectores(v1, v2))
    else:
        print("Los vectores deben tener la misma dimensión")

except ValueError:
    print("Debe ingresar solo números.")
```
