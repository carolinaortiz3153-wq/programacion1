```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()
    password = input("Contraseña: ")

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    if usuarios[usuario] != password:
        raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

```python
try:
    num1 = float(input("numero 1: "))
    num2 = float(input("numero 2: "))
    op = input("operacion (+, -, *, /): ")

    if op == "+":
      resultado = num1 + num2
    elif op == "-":
      resultado = num1 - num2
    elif op == "*":
      resultado = num1 * num2
    elif op == "/":
      resultado = num1 / num2
    else:
      raise ValueError("operacion no valida")
    print("resultado", resultado)
except TypeError as e:
    print("Error de tipo: ", e)
```

```python
productos = {
    "pan": 1.5,
    "leche": 2.0,
    "huevos": 3.0
}
print("Productos disponibles:", productos)

try:
  nombre = input("¿Qué producto quieres?: ").lower()
  if nombre not in productos:
    raise KeyError("El producto no está disponible")

  cantidad = int(input("Cantidad: "))
  if cantidad <= 0:
    raise ValueError("La cantidad debe ser mayor a 0")
except KeyError as e:
  print("Error:", e)
except ValueError as e:
  print("Error:", e)
else:
 total = productos[nombre] * cantidad
 print("Total a pagar:", total)
```

```python
try:
    datos = input("Ingresa notas separadas por coma: ")

    notas = [float for n in datos.split(",")]

    if len(notas) == 0:
        raise ValueError("No ingresaste notas")

    promedio = sum(notas) / len(notas)

    posicion = int(input("¿Qué nota quieres ver (índice)?: "))

    if posicion < 0:
        raise ValueError("Ingresa solo índices positivos")

    print("Nota seleccionada:", notas[posicion])

    print("Promedio:", promedio)

except ValueError as e:
    print("Error de valor:", e)

except IndexError:
    print("Índice fuera de rango")

except TypeError:
    print("Error: tipos de datos no compatibles")
```

el metodo findall muestra los numeros que aparecen en el texto incluyendo cadenas de numeros que son mostradas digito por digito

```python
import re
texto = "tengo 2 gatos y 12 pistolas"
resultado = re.findall(r"\d+", texto)
print(resultado)
```
