```python
import numpy as np
import pandas as pd

np.random.seed(123) #para guardar la semilla
```

```python
datos = np.random.normal(loc=4.0, scale=0.5, size=50)
#simula 50 valores de la normal con media 4 y desviacion estandar 0.5
df = pd.DataFrame({
    "notas" : datos
})
```

```python
df.describe()
```

aproximadamente el 75% de los datos (notas) son hasta 4.4

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Estilo
sns.set_theme(style="whitegrid")

# Datos limpios
data = df["notas"].dropna().values

# Ajuste normal
mu, std = norm.fit(data)

# Rango para la curva
x = np.linspace(data.min(), data.max(), 200)

# PDF normal
pdf = norm.pdf(x, mu, std)

# Figura
plt.figure(figsize=(10, 6))

# Histograma (normalizado correctamente)
plt.hist(data, bins=20, density=True, alpha=0.5,
         color="skyblue", edgecolor="black", label="Histograma")

# KDE (más estable así)
sns.kdeplot(x=data, color="navy", linewidth=2, label="KDE (Kernel)")

# Normal ajustada
plt.plot(x, pdf, 'r--', linewidth=2,
         label=f'Normal ajustada (μ={mu:.2f}, σ={std:.2f})')

# Detalles
plt.title("Distribución de Notas", fontsize=16)
plt.xlabel("Notas")
plt.ylabel("Densidad")
plt.legend()
plt.tight_layout()

plt.show()
```

```python
from scipy.stats import shapiro
shapiro(df["notas"])
```

si el valor p es mayor o igul a 0.05 no se rechaza la hipotesis de normalidad con la una confianza 95%

```python
import numpy as np
import pandas as pd
from scipy import stats

np.random.seed(123)

grupo_A = np.random.normal(loc=4.0, scale=0.4, size=40)
grupo_B = np.random.normal(loc=4.3, scale=0.5, size=40)

df = pd.DataFrame({
    "A": grupo_A,
    "B": grupo_B
})

t_stat, p_val = stats.ttest_ind(df["A"], df["B"])

print("Estadístico t:", t_stat)
print("Valor p:", p_val)

```

se rechaza que las calificaciones sean iguales para ambos grupos con una confianza del 95% ya que el valor p es menor a 0.05
