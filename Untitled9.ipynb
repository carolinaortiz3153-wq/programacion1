```python
def sumar_matrices(A, B):
  filas = len(A)
  columnas = len(A[0])

  if filas != len(B) or columnas != len(B[0]):
    return "Error: Las matrices deben tener las mismas dimensiones para sumarse."

  else:
    resultado = []
    for i in range(filas):
      fila = []
      for j in range(columnas):
        fila.append(A[i][j] + B[i][j])
        resultado.append(fila)
  return resultado
```

```python
filas = int(input("Ingrese el número de filas: "))

def matriz(nombre):
    matriz = []
    for i in range(filas):
        fila = list(map(float, input(f"Ingrese la fila {i+1} de {nombre} separada por espacios: ").split()))
        matriz.append(fila)
    return matriz


M3 = matriz("M3")
M4 = matriz("M4")


def producto_hadamard1(M3, M4):
    filas = len(M3)
    columnas = len(M3[0])

    if filas != len(M4) or columnas != len(M4[0]):
        return "Error: Las matrices deben tener las mismas dimensiones."

    try:
        resultado = []
        for i in range(filas):
            fila = []
            for j in range(columnas):
                fila.append(M3[i][j] * M4[i][j])
            resultado.append(fila)
        return resultado

    except ValueError:
        return "Debe ingresar solo números."


resultado = producto_hadamard1(M3, M4)

print("Resultado:")
for fila in resultado:
    print(fila)
```
