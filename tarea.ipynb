```python
# lista
verdes = ["phtalo", "vejiga", "permanente"]
tupla = tuple(verdes)
diccionario = {"phtalo": "perfecto",
               "vejiga": "medio util",
               "permanente": "meh"}
set = (verdes)
print(verdes)
print(tupla)
print(diccionario)
print(set)
```

```python
# tupla
azules = ("prusia", "cobalto", "ceruleo")
lista = [azules]
diccionario = (azules)
set = (azules)
print(azules)
print(lista)
print(diccionario)
print(set)
```

```python
# diccionario
amarillos = {
    "ocre": "feo",
    "siena natural": "util",
    "cadmio": "lindo"}
lista = [amarillos]
tupla = tuple(amarillos)
set = (amarillos)
print(amarillos)
print(lista)
print(tupla)
print(set)
```

```python
# conjunto
rojos = {"cadmio", "permanente", "carmin"}
lista = [rojos]
tupla = tuple(rojos)
diccionario = {"cadmio": "util",
               "permanente": "primario",
               "carmin": "lindo"}
print(rojos)
print(lista)
print(tupla)
print(diccionario)
```
